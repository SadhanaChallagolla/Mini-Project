## Importing neccesary Libraries

## Preprocessing

```python
for col in df.columns:
    print(col, ':', len(df[col].unique()),'labels')
    
    

pd.get_dummies(df, drop_first=True).shape
```

```python
pd.get_dummies(df, drop_first=True).shape
```

```python
symptoms_1 = [x for x in df.Symptom_1.value_counts().sort_values(ascending=False).head(34).index]
symptoms_2 = [x for x in df.Symptom_2.value_counts().sort_values(ascending=False).head(40).index]
symptoms_3 = [x for x in df.Symptom_3.value_counts().sort_values(ascending=False).head(54).index]
symptoms_4 = [x for x in df.Symptom_4.value_counts().sort_values(ascending=False).head(50).index]
#symptoms_5 = [x for x in df.Symptom_5.value_counts().sort_values(ascending=False).head(35).index]
```

```python
symptoms = symptoms_1 + symptoms_2 + symptoms_3 + symptoms_4

def unique(symptoms):
    unique_list = []
    
    for x in symptoms:
        if x not in unique_list:
            unique_list.append(x)
    return unique_list

symptoms = unique(symptoms)
```

```python
for label in symptoms_1:
    df[label] = np.where(df['Symptom_1'] == label, 1, 0)
    
for label in symptoms_2:
    df[label] = np.where(df['Symptom_2'] == label, 1, 0)
    

for label in symptoms_3:
    df[label] = np.where(df['Symptom_3'] == label, 1, 0)
    
for label in symptoms_4:
    df[label] = np.where(df['Symptom_4'] == label, 1, 0)
    
#for label in symptoms_5:
    #df[label] = np.where(df['Symptom_5'] == label, 1, 0)      
dt = df[['Disease']+ symptoms ]
dt
```

```python
labelencoder = LabelEncoder()
dt['Disease']= labelencoder.fit_transform(dt['Disease'])

y_encode_disease = dt['Disease']
dt
```

```python
y_encode_disease_unique = y_encode_disease.unique()
y_encode_disease_unique
```

```python
Y= dt['Disease']
X= dt.drop('Disease', axis=1)
```

### Standard feature scaling

```python
ss = StandardScaler()
X = ss.fit_transform(X)
```

## Train_Test_split

```python
x_train,x_test,y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state =50)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Decision Tree Classifier

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

```python
dtree_pred = dtree.predict(x_test)
```

```python
dtree_pred, dtree_pred.shape
```

```python
accuracy_score(y_test, dtree_pred)
```

```python
confusion_matrix(y_test, dtree_pred)
```

## Random Forest classifier

```python
rfc = RandomForestClassifier(n_estimators=300)
rfc.fit(x_train, y_train)
```

```python
rfc_pred = rfc.predict(x_test)
```

```python
rfc_pred
```

```python
accuracy_score(y_test, rfc_pred)
```

```python
confusion_matrix(y_test, rfc_pred)
```

## Bagging Classifier

```python
bg = BaggingClassifier(DecisionTreeClassifier(), max_samples=0.5, max_features = 1.0, n_estimators=25)
bg.fit(x_train, y_train)
```

```python
bg_pred = bg.predict(x_test)
bg_pred
```

```python
bg.score(x_test, y_test)
```

```python
bg.score(y_test, bg_pred)
```

## Boosting Classifier

```python
ada =AdaBoostClassifier(DecisionTreeClassifier(), n_estimators=10, learning_rate=1)
ada.fit(x_train, y_train)
```

```python
ada.predict(x_test)
```

```python
ada.score(x_test, y_test)
```

```python
def Take_input():
    inp = input_text.get("1.0", "end-1c")
    #print(inp)
    list1= []
    list1 = [x for x in inp.split(',')]
    #print(list1)
    list2 = []
    
    for x in symptoms:
        if x in list1:
            list2.append(1)
        else:
            list2.append(0)
    
    #print(list2)
    dis_pred= dtree.predict([list2])
    #print(dis_pred)
    
    y_disease_list = y_encode_disease_unique.tolist()
    #print(y_disease_list)
    y_dis = y_disease_list.index(dis_pred[0])
    #print(y_dis)
    #print(Y_dt[y_dis])
    
    Label(screen, text=Y_dt[y_dis], font={"calibri", 16}).pack()
```

```python
def main_screen():
    global screen
    screen = Tk()
    
    global input_text
    screen.geometry("300x250")
    
    screen.title("Disease prediction by symptoms")
    Label(text="Disease Prediction by Symptoms", width="200",height="5",fg="yellow",bg="grey" ,font={"calibri", 100}).pack()
    Label(text="").pack()
    
    img = Image.open(r"C:\Users\chsad\Documents\Mini project\Prescription.jpg")
    test=ImageTk.PhotoImage(img)
    label1 = Label(image =test)
    label1.image = test
    label1.place(x=30, y=120) #x=30 y=120
    
    
    Label(text="Symptoms:", width="30",height="2",fg="Black", font = {"Times new roman", 16}).pack()
    input_text = Text(screen, height=10, width=35)
    input_text.pack()
    Label(text="").pack()
    
    Button(screen, text="Submit", width="15", height="2", command = Take_input).pack()
    Label(text="").pack()
    screen.mainloop()
```
